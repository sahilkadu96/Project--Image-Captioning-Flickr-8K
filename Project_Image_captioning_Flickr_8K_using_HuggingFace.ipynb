```python
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
#Salesforce blip image captioning base is a model from HugingFace
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
```

```python
img_url = '/content/drive/MyDrive/Flickr 8K images/111766423_4522d36e56.jpg' 
raw_image = Image.open(img_url).convert('RGB')
raw_image
```

```python
inputs = processor(raw_image, return_tensors="pt")

out = model.generate(**inputs)
print(processor.decode(out[0], skip_special_tokens=True))
```

```python
def generate_image_caption(image_url):
  raw_image = Image.open(image_url).convert('RGB')
  inputs = processor(raw_image, return_tensors="pt")
  out = model.generate(**inputs)
  caption = processor.decode(out[0], skip_special_tokens=True)
  img_array = mpimg.imread(image_url)
  plt.imshow(img_array)
  return caption
  
```
